# Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('zomato.csv')
df
```

# Cleaning the data

```python
df.shape
```

```python
df.columns
```

```python
df = df.drop(['url', 'address', 'phone', 'menu_item', 'dish_liked', 'reviews_list'], axis=1)
df
```

```python
df.info()
```

```python
df.drop_duplicates(inplace=True)
df.shape
```

### Cleaning 'rate' column

```python
df['rate'].unique()
```

```python
def clean_rate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)
    
df['rate'] = df['rate'].apply(clean_rate)
df['rate']
```

```python
# Let's find number of null values in the column 'Rate'
df.rate.isnull().sum()
```

Let's fill these null values in rate column with mean

```python
df['rate'].fillna(df['rate'].mean(), inplace = True)
df['rate'].isnull().sum()
```

```python
df.info()
```

### Let's drop the null values for the rest of columns

```python
df.dropna(inplace = True)
df.info()
```

```python
df.rename(columns = {'approx_cost(for two people)':'Cost2plate', 'listed_in(type)':'Type', 'listed_in(city)':'City'}, inplace = True)
df.head()
```

```python
df['location'].unique()
```

```python
df['City'].unique()
```

```python
#Let's drop one of them since both are similar columns
df = df.drop(['City'], axis=1)
```

```python
df.head()
```

### Cleaning Cost2plate column

```python
df['Cost2plate'].unique()
```

```python
#Let's convert these values from string to float

def convert_to_float(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',', '')
        return float(value)
    else:
        return float(value)
    
df['Cost2plate'] = df['Cost2plate'].apply(convert_to_float)
df['Cost2plate'].unique()
```

```python
df.head()
```

### Cleaning 'rest_type' column 

```python
rest_types = df['rest_type'].value_counts(ascending = False)
rest_types
```

```python
rest_types_lessthan1000 = rest_types[rest_types<1000]
rest_types_lessthan1000
```

```python
def merging_rest_types(value):
    if(value in rest_types_lessthan1000):
        return 'Others'
    else:
        return value
    
df['rest_type'] = df['rest_type'].apply(merging_rest_types)
df['rest_type'].value_counts()
```

### Cleaning 'Location' column

```python
df['location'].value_counts()
```

```python
location = df['location'].value_counts(ascending = False)

location_lessthan300 = location[location<300]
location_lessthan300
```

```python
def merge_location(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value
    
df['location'] = df['location'].apply(merge_location)
df['location'].value_counts()
```

```python
df.head()
```

### Cleaning 'cuisines' column

```python
cuisines = df['cuisines'].value_counts(ascending = False)
cuisines
```

```python
cuisines_lessthan100 = cuisines[cuisines<100]
cuisines_lessthan100
```

```python
def merge_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'others'
    else:
        return value
    
df['cuisines'] = df['cuisines'].apply(merge_cuisines)
df['cuisines'].value_counts()
```

```python
df.head()
```

```python
zomato_cleaned_dataset = df
zomato_cleaned_dataset.to_csv('zomato_cleaned_dataset.csv')
zomato_cleaned_dataset = pd.read_csv('zomato_cleaned_dataset.csv')
zomato_cleaned_dataset
```

## Visualising online order facility, location wise

```python
df1 = df.groupby(['location', 'online_order'])['name'].count()
df1.to_csv('location_online.csv')
df1 = pd.read_csv('location_online.csv')
df1
```

```python
df2 = pd.pivot_table(df1, values=None, index=['location'], columns=['online_order'], fill_value=0, aggfunc=np.sum)
df2
```

```python
df2.plot(kind='bar', figsize=(15,8))
```

```python
plt.figure(figsize=(14,8))
sns.boxplot(x='Type', y='rate', data=df)
```

## Grouping Types of Restaurants, location wise

```python
df3 = df.groupby(['location', 'Type'])['name'].count()
df3.to_csv('location_Type.csv')
df3 = pd.read_csv('location_Type.csv')
df3
```

```python
df4 = pd.pivot_table(df3, values=None, index=['location'], columns=['Type'], fill_value=0, aggfunc=np.sum)
df4
```

```python
df4.plot(kind='bar', figsize=(36,8))
```

## No. of Votes, Location Wise

```python
df5 = df[['location', 'votes']]
df5.drop_duplicates()
df6 = df5.groupby(['location'])['votes'].sum()
df6 = df6.to_frame()
df6 = df6.sort_values('votes', ascending=False)
df6
```

```python
plt.figure(figsize=(15, 8))
sns.barplot(x=df6.index, y=df6['votes'])
plt.xticks(rotation=90)
plt.show()
```
